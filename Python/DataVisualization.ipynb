```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'], df_sales['Fridge'],label="Fridge")
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'],label="Dishwasher")
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'],label="Washing Machine")
plt.title("Product Sales")
plt.xlabel("Quarter")
plt.ylabel("Revenue (in millions)")
plt.legend()
plt.show()
```

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.2f",explode=[0.1,0,0],shadow=True)
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter",xlabel="Quarter",ylabel="Revenue (in millions)",title="Sales by Quarter")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales2 = df_sales.set_index("Quarter")
df_sales2.plot(kind="bar",xlabel="Quarter",ylabel="Revenue (in millions)",title="Sales by Quarter")
plt.xticks(rotation=45)
plt.show()
```
