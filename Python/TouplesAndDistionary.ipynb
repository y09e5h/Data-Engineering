## Touples


```python
def sumAndMull(*args):
    sum = 0
    mul = 1
    for num in args:
        sum += num
        mul *= num
    return sum,mul

result= sumAndMull(1,23,4,2,6,4)
print(type(result))
print(result)
```

## Dictionary

```python
contact_info={
    "Yogesh" : {"phone": 123,"addr":"Sangli"},
    "Namrata" : {"phone": 456,"addr":"Nashik"}
}

```

```python
contact_info.get("Yogesh")
```

```python


contact_info["Ramesh"]= {"phone": 999,"addr":"Phulawadi"}
```

```python
for name,addr in contact_info.items():
    print(name," : ")
    for k,v in addr.items():
        print("\t",k,v)
```
