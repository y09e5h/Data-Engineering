```python
def curr_convert(curr):
    if curr in ["$$","Dollars"]:
        return "USD"
    else:
        return curr
df_financials = pd.read_excel("movies_db.xlsx","financials",converters={"currency":curr_convert})
df_financials.head()
```

```python
df_merged=pd.merge(df_movies,df_financials,on="movie_id",how="inner")
```

```python

df_movies.merge(df_financials,on="movie_id",how="outer")
```

```python
df_movies.head()
```

```python
df_merged.to_excel("mearged.xlsx", sheet_name="merged",index=False)
```
