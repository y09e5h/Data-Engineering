```python
player_data={}
with open("score.csv","r") as file:
    for line in file:
        name,_,score = line.split(",")
        if not score.strip().isdigit():
            continue
        score = int(score.strip())
        if name not in player_data:
            player_data[name]=[score]
        else:
            player_data[name].append(score)
player_data
```

```python
for player,scorelist in player_data.items():
    print(f"Player {player} -> Min {min(scorelist)}, Max {max(scorelist)}, Avg {round(sum(scorelist)/len(scorelist),2)} ")
```
