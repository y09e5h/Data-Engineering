### Total Function
To avoid code redundancy

```python
yogesh_expenses=[100,20,70,400,50,2,8,11,450]
Namrata_expenses=[12000,2400,1200,700,490,4500,1100,570]

def total_expenses(expenses):
    '''
    :param expenses: input list containing numbers
    :return: sum of all expenses
    '''
    return sum(expenses)

print("Yogesh expenses:",total_expenses(yogesh_expenses))
print("Namrata expenses:",total_expenses(Namrata_expenses))
```

### find volume of cylinder

```python
import math
def cylinder_volume(r,h=10):#default hight is 10
    return math.pi*(r**2)*h

r=10
h=5
print(f"Volume for R= {r} and H = {h} = ",cylinder_volume(r,h))
print(f"Volume for R= {r} and H = default = ",cylinder_volume(r))
```

### Function with any number of arguments
args and kwargs


```python
def sum_all(*args):#args
    total=0
    for arg in args:
        total+=arg
    return total

print("SUM : ",sum_all(10,23,43,23,2,34))
```

```python
def company_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")

company_info(ticker='AAPl',ceo="Tim Cook",revenue="200 billion",pe=20,pb=10.2)
```
