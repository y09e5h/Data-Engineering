### Common Words Between Two Sentences

```python
from collections import Counter
import pandas as pd
import string
def find_common_words(row):
    translator = str.maketrans('', '', string.punctuation)
    s1 = set(row['sentence_1'].translate(translator).split())
    s2 = set(row['sentence_2'].translate(translator).split())
    common = s1 & s2
    counts = Counter(s2)
    return {word : counts[word] for word in sorted(common)}

    # Your logic here

df = pd.read_csv("data.csv")
df['output'] = df.apply(find_common_words, axis=1)

df

```
