### Palindrome Check

```python
import pandas as pd
import string
df = pd.read_csv("data.csv")
def is_palindrome(row):
    translator = str.maketrans('','',string.punctuation)
    s = row["text"].translate(translator).lower().replace(" ","")
    return s == s[::-1]
df["is_palindrome"] = df.apply(lambda row: is_palindrome(row),axis=1)
df
# return the DataFrame
#df

```
